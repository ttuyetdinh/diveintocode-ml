# Problem 1

```
import numpy as np
mean=np.array([-3,0])
cov=np.array([[1.0,0.8],[0.8,1.0]])
np.random.seed(0)
x=np.random.multivariate_normal(mean,cov,(500))
print(x.shape)
```

#Problem 2

```
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(x[:,0], x[:,1])
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("scatter")
plt.show()
```

# Problem 3


```
import matplotlib.pyplot as plt
%matplotlib inline
plt.hist(x[:,0],50)
plt.xlabel("x value")
plt.ylabel("Frequency")
plt.xlim(-6,2)
plt.title("Histogram of x1")
plt.show()
```

```
import matplotlib.pyplot as plt
%matplotlib inline
plt.hist(x[:,1],50)
plt.xlabel("x value")
plt.ylabel("Frequency")
plt.xlim(-6,2)
plt.title("Histogram of x2")
plt.show()
```

# Problem 4

```
mean1=np.array([0,-3])
y=np.random.multivariate_normal(mean1,cov,(500))
plt.scatter(x[:,0], x[:,1], label="0")
plt.scatter(y[:,0], y[:,1], label="1")
plt.xlabel("x value")
plt.ylabel("Frequency")
plt.title("scatter")
plt.legend()
plt.show()
```

#Problem 5

```
t=np.concatenate((x,y),axis=0)
print(t.shape)
print(t)
```

# Problem 6

```
x_temp=np.full((500,1), 0)
y_temp=np.full((500,1), 1)
x_temp=np.concatenate((x,x_temp),axis=1)
y_temp=np.concatenate((y,y_temp),axis=1)
t=np.concatenate((x_temp,y_temp),axis=0)
print(t.shape)
print(t)
```
