#Problem
According to cookpad recipe the chestnut buns is bigger than human palm.To simplify, its shaped almost like a circle with r=5cm so it acreage is 78 cm2= 7.8e-3 m2.

  We treat solar system as a circle too, with r=5.914e12 m, so it acreage is 1.099e26 m2


[https://cookpad.com/uk/recipes/155252-chestnut-buns](https://)
photo.png

```
import math
import numpy as np
import matplotlib.pyplot as plt
SOLAR_SIZE = 1.099e26
CHESTNUT_BUNS = 7.8e-3
total_cake=[]
solar_size=[]
def sprinkle(sizebig=1023,sizetiny=1):
  list1=[]
  list2=[]
  timeth=0;
  cake=sizetiny
  totalcake=1
  while(totalcake-cake<sizebig):
    list1.append(sizebig)
    list2.append(totalcake)
    timeth+=1
    cake*=2
    totalcake+=cake
  return list1,list2,timeth  
solar_size,total_cake,min_left= sprinkle(SOLAR_SIZE, CHESTNUT_BUNS)
min_left*=5
print("Minutes to fill the solar system: {} [min]".format(min_left))
%matplotlib inline
plt.title("Fill the solar system")
plt.xlabel("Minutes")
plt.ylabel("Size")
x=np.arange(0,min_left,5)
plt.plot(x, total_cake, label="Size of  total cake") 
plt.plot(x, solar_size, label="Size of solar system") 
plt.legend()
plt.show() 
```
