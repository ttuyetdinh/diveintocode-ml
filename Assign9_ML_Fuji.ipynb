```python
import numpy as np
from google.colab import drive
import matplotlib.pyplot as plt
drive.mount('/content/drive')

csv_path = "/content/drive/MyDrive/Colab Notebooks/mtfuji_data.csv" # Specify the file name (path)

np.set_printoptions(suppress=True) # Set the prohibition of exponential notation
fuji = np.loadtxt(csv_path, delimiter=",", skiprows=1)
plt.title("Mt.Fuji")
plt.xlabel("position")
plt.xlim(fuji[0:, 0:1].min()*1.1,fuji[0:, 0:1].max()*1.1)
#plt.ylim(gradient.min()*1.1,gradient.max()*1e-12)
plt.ylabel("elevatoin(m)")
plt.plot(fuji[0:, 0:1], fuji[0:, 3:4], label="meters")
plt.legend()
plt.show()   
```

# Problem 2

```python
def cal_gradient(data_csv):
  d_number=data_csv[1:, 0:1]-data_csv[:-1, 0:1]
  d_elevation=data_csv[1:, 3:4]-data_csv[:-1, 3:4]
  gradient=d_elevation/d_number
  return gradient
gradient_arr=cal_gradient(fuji)
print(gradient_arr.shape)
```

#Problem 3

```python
def des_heights(fuji_csv, gradient, curr_pt=0, orgi_pt=0, alpha=0.2):
  des_pt=curr_pt
  des_pt=des_pt-alpha*gradient[curr_pt][0]
  return round(des_pt)
k=des_height(fuji, gradient_arr, 136,136)
print(k)
```

#Problem 4

```python
def go_downs(fuji_csv, gradient,orgi_pt=136,alpha=0.2):
  trackheight=[]
  trackpoint=[]
  curr_pt=orgi_pt
  i=0
  while(curr_pt>=0 and curr_pt<300 and i<1000):
    trackpoint.append(curr_pt)
    trackheight.append(fuji_csv[curr_pt][3])
    curr_pt=des_heights(fuji_csv,gradient,curr_pt,orgi_pt,alpha)
    i+=1
  return trackpoint,trackheight
trackpoint_lists,trackheight_lists= go_downs(fuji,gradient_arr,136,0.2)
print(trackpoint_lists)
print(trackheight_lists)
```

#Problem 5

```python
def plot_graph(x,y,fuji):  
  plt.title("Mt.Fuji")
  plt.xlabel("Position")
  plt.xlim(fuji[0:, 0:1].min()*1.1,fuji[0:, 0:1].max()*1.1)
  #plt.ylim(gradient.min()*1.1,gradient.max()*1e-12)
  plt.ylabel("Elevation(m)")
  plt.plot(fuji[0:, 0:1], fuji[0:, 3:4], label="Real data")
  plt.plot(x,y, label="Machine predict")
  plt.legend()
  plt.show()    
plot_graph(trackpoint_lists,trackheight_lists,fuji)  
```
