#Problem 1

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
data_read=load_iris()
x=pd.DataFrame(data = data_read.data, columns=data_read.feature_names)
print(x)
y=pd.DataFrame(data_read.target, columns=['Species'])
print(y)
print(data_read.target_names)
```

#Problem 2

```python
df= pd.concat([x,y],axis=1)
print(df)
```

#Problem 3

```python
print("First 4 sample of the dataset:")
print(df.head(4))
print("\nInfo of the dataset:")
df.info()
df.describe()
```

#Problem 6

```python
pie=df['Species'].value_counts()
plt.pie(pie, labels=data_read.target_names, explode = (0.1, 0, 0), autopct='%1.2f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

```python
# visualize the distribution of data for each label using a box plot
sns.set_theme(style = "whitegrid")
fig1, axes = plt.subplots(3, 1, figsize = (10,20))
axes[0].set_title("setosa")
sns.boxplot(ax = axes[0], data = df.iloc[:50,0:4])
axes[1].set_title("versicolor")
sns.boxplot(ax = axes[1], data = df.iloc[50:100,0:4])
axes[2].set_title("virginica")
sns.boxplot(ax = axes[2], data = df.iloc[100:150,0:4])
```

```python
# visualize the distribution of data for each label using a violin plot
fig1, axes = plt.subplots(3, 1, figsize = (10,20))
axes[0].set_title("setosa")
sns.violinplot(ax = axes[0], data = df.iloc[:50,0:4])
axes[1].set_title("versicolor")
sns.violinplot(ax = axes[1], data = df.iloc[50:100,0:4])
axes[2].set_title("virginica")
sns.violinplot(ax = axes[2], data = df.iloc[100:150,0:4])
```

* Box plot: shows “four main features about a variable: center, spread, asymmetry, and outliers”
* Violin plot: combines the box plot and the density trace (or smoothed histogram) into a single display that reveals structure found within the data
